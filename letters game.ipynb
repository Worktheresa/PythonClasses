```python
def level_one():
    words = []
    six_letter_words = []
    total_inputs = 5

    while len(words) < total_inputs:
        word = input(f"Enter word ({len(words) + 1}/{total_inputs}): ")
        words.append(word)
        
        if len(word) == 6:
            six_letter_words.append(word)
            print("This is a six-letter word.")
        elif len(word) < 6:
            print("This word has less than six letters.")
        else:
            print("This word has more than six letters.")
        
        print(f"You have {total_inputs - len(words)} inputs left.")
    
    print("\nSix-letter words entered:", six_letter_words)
    return six_letter_words

def level_two():
    words = []
    seven_letter_words = []
    total_inputs = 4

    while len(words) < total_inputs:
        word = input(f"Enter word ({len(words) + 1}/{total_inputs}): ")
        words.append(word)
        
        if len(word) == 7:
            seven_letter_words.append(word)
            print("This is a seven-letter word.")
        elif len(word) < 7:
            print("This word has less than seven letters.")
        else:
            print("This word has more than seven letters.")
        
        print(f"You have {total_inputs - len(words)} inputs left.")
    
    print("\nSeven-letter words entered:", seven_letter_words)
    return seven_letter_words

def level_three():
    words = []
    nine_letter_words = []
    total_inputs = 3

    while len(words) < total_inputs:
        word = input(f"Enter word ({len(words) + 1}/{total_inputs}): ")
        words.append(word)
        
        if len(word) == 9:
            nine_letter_words.append(word)
            print("This is a nine-letter word.")
        elif len(word) < 9:
            print("This word has less than nine letters.")
        else:
            print("This word has more than nine letters.")
        
        print(f"You have {total_inputs - len(words)} inputs left.")
    
    print("\nNine-letter words entered:", nine_letter_words)
    return nine_letter_words

def main():
    print("Level 1: Six-letter words with 5 trials")
    level_one_words = level_one()
    
    print("\nLevel 2: Seven-letter words with 4 trials")
    level_two_words = level_two()
    
    print("\nLevel 3: Nine-letter words with 3 trials")
    level_three_words = level_three()
    
    print("\nSummary:")
    print("Six-letter words entered:", level_one_words)
    print("Seven-letter words entered:", level_two_words)
    print("Nine-letter words entered:", level_three_words)

if __name__ == "__main__":
    main()
```
