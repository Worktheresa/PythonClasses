```python
#          ScheduleATask GUI application

# Programmablity
import datetime, os, tkinter as tk
from tkinter import messagebox


class ScheduleTask:
    def __init__(self):
        
        #    GUI
        #Creating the application's window
        self.GUI = tk.Tk()
        self.GUI.geometry('600x500')
        
        #Adding a heading and setting it's geometry
        self.Header = tk.Label(self.GUI, text = "SHCEDULE A TASK!", bg = "gray",font = ("Calibre", 15, "bold") )
        self.Header.place(x = 200, y = 0)

        # Adding the widgets for scheduling a task
        #Widgets on the left
        self.Task_Label = tk.Label(self.GUI, text = "Task:", font = ("Calibre", 10, "bold"))
        self.Task_Label.place(x = 40, y = 60)
        self.Task_entry = tk.Entry(self.GUI)
        self.Task_entry.place(x = 40, y = 80)
        
        self.Year_Label = tk.Label(self.GUI, text = "Year:", font = ("Calibre", 10, "bold"))
        self.Year_Label.place(x = 40, y = 120)
        self.Year_entry = tk.Entry(self.GUI)
        self.Year_entry.place(x = 40, y = 140)

        self.Month_Label = tk.Label(self.GUI, text = "Month:", font = ("Calibre", 10, "bold"))
        self.Month_Label.place(x = 40, y = 180)
        self.Month_entry = tk.Entry(self.GUI)
        self.Month_entry.place(x = 40, y = 200)

        self.Day_Label = tk.Label(self.GUI, text = "Day:", font = ("Calibre", 10, "bold"))
        self.Day_Label.place(x = 40, y = 240)
        self.Day_entry = tk.Entry(self.GUI)
        self.Day_entry.place(x = 40, y = 260)

        self.Hour_Label = tk.Label(self.GUI, text = "Hour:", font = ("Calibre", 10, "bold"))
        self.Hour_Label.place(x = 40, y = 300)
        self.Hour_entry = tk.Entry(self.GUI)
        self.Hour_entry.place(x = 40, y = 320)

        # The task submit button
        self.button_schedule = tk.Button(self.GUI, text = "Create Schedule", bg= "grey", command = self.buttonScheduleTask)
        self.button_schedule.place(x = 70, y = 360)

        #Widgets on the rigth
        self.button_viewTask = tk.Button(text = "View tasks", bg = "grey", command = self.buttonViewTask)
        self.button_viewTask.place(x = 450, y = 70)

        self.frame1 = tk.Frame(self.GUI, height = 250, width= 150, bd = 10, bg = 'white')
        self.frame1.place(x = 400, y = 120)
        self.viewTask_label = tk.Label(self.frame1, text = "Pending")
        self.viewTask_label.place(x = 4, y = 2)

        #        Programs
        #Creating the storage file
        self.txt_file = open("Scheduled_tasks.txt", 'wt')
        

        # Running the GUI
        self.GUI.mainloop()
        
    


    def buttonScheduleTask(self):
        # Program's Body
        schedule_date = None
        self.user_data = {
             "Task" : self.Task_entry.get(),
             "Year" : self.Year_entry.get(),
             "Month" : self.Month_entry.get(),
             "Day" : self.Day_entry.get(),
             "Hour" : self.Hour_entry.get()
        }
        while True:
            # Creating the date object, and validating it
            try:
                self.schedule_date = datetime.datetime(int(self.user_data.get("Year"))  , int(self.user_data.get("Month")), int(self.user_data.get("Day")), int(self.user_data.get("Hour")))
            except:
                messagebox.showinfo("Error", "Please ensure that the fields are not empty, and the rigth data are being passed")
                break

            # Validating if the schedule is behind time
            if( self.schedule_date < datetime.datetime.now()):
                messagebox.showinfo("Error", "Sorry, your schedule is behind the current time. Try again")
                break

            # Storing the task and it's date in a text file
            self.txt_file.write(f'TASK :\t\t{self.user_data.get("Task")}\nYEAR :\t\t{self.schedule_date.year}\nMONTH :\t\t{self.schedule_date.month}\nDAY :\t\t{self.schedule_date.day}\nHOUR :\t\t{self.schedule_date.hour}\n\n')

            # Displaying the task and it's approved date
            messagebox.showinfo("Success", "Your task has been scheduled successfully!")
            #print(f'\n\nInputs successful\nTASK : {self.user_data.get("Task")}\nYEAR : {self.schedule_date.year}\nMONTH : {self.schedule_date.month}\nDay : {self.schedule_date.day}\nHOUR : {self.schedule_date.hour}')
            break


    def buttonViewTask(self):
        # Closing the previous connection
        self.txt_file.close()
        # Creating a new connection
        self.txt_file = open("Scheduled_tasks.txt")
        file_data = self.txt_file.read()
        # Checking if the file is empty
        if len(file_data) < 1:
            self.viewTask_label.config(text = f'No schedule yet!')
        else:
            self.viewTask_label.config(text = f'TODO List\n{file_data}')
        
        # Closing this connection
        self.txt_file.close()
        # Opening the file back in append mode
        self.txt_file = open("Scheduled_tasks.txt", "at")
        
    def close(self):
        self.txt_file.close()

shc_obj = ScheduleTask()
```
